#### Filling missing values

```python
df.fillna('Not Defined')
```

#### Filling missing values with the mean values of the column

```python
df['UnitPrice'] = df['UnitPrice'].fillna(df['UnitPrice'].mean())
df
```

### Renaming Columns

```python
df = df.rename(columns={'OrderDate':'Date'})
df.head()
```

#### Changing Datatypes

```python
df['UnitPrice_new'] = df['UnitPrice'].fillna(df['UnitPrice'].mean()).astype(int)
df.head()
```

### Adding new Column

```python
df['New UnitPrice'] = df['UnitPrice'].apply(lambda x : x * 2)
df.head()
```

### Data Aggregation and Grouping

```python
grouped_mean = df.groupby('Product')['UnitPrice'].mean()
grouped_mean
```

```python
df.groupby(['Product', 'City'])['UnitPrice'].sum()
```

#### Aggregate Multiple functions

```python
df.groupby('City')['UnitPrice'].agg(['mean', 'sum', 'count'])
```

#### Merging and Joining DataFrames

##### Creating Sample DataFrames

```python
df1 = pd.DataFrame({'Key' : ['A', 'B', 'C'], 'Value1' : [1, 2, 3]})
df2 = pd.DataFrame({'Key' : ['A', 'B', 'D'], 'Value2' : [4, 5, 6]})
```

```python
df1
```

```python
df2
```

##### Merge DataFrames on 'Key' Columns

_inner_ join: only common keys will be merged from both DataFrames

```python
pd.merge(df1, df2, on='Key', how='inner')
```

_outer_ join: all keys will be merged from both DataFrames

```python
pd.merge(df1, df2, on='Key', how='outer')
```

_left_ outer join: first DataFrame will be given more priority in terms of joining, i.e., all keys of left DataFrame will be merged with respect to right Dataframe

```python
pd.merge(df1, df2, on='Key', how='left')
```

_right_ outer join: second DataFrame will be given more priority in terms of joining, i.e., all keys of right DataFrame will be merged with respect to left Dataframe

```python
pd.merge(df1, df2, on='Key', how='right')
```
