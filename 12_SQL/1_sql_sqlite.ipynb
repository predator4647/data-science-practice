# SQL and SQLite
SQL (Structured Query Language) is a standard language for managing and manipulating relational databases. SQLite is a self-contained, severless, and zero-configuration database engine that is widely used for embedded database systems. In this lesson, we will cover the basics of SQL and SQLite, including creating databases, tables and performing various SQL operations.

```python
import sqlite3
```

### Connect to a SQLite Database

```python
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
cursor
```

### Create a Table

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS EMPLOYEES(
    ID INTEGER PRIMARY KEY,
    NAME VARCHAR NOT NULL,
    AGE INTEGER,
    DEPARTMENT VARCHAR
)
""")

## Commit the changes
connection.commit()
```

```python
cursor.execute("SELECT * FROM EMPLOYEES")
```

### Insert Values

```python
cursor.execute(
    "INSERT INTO EMPLOYEES VALUES(1, 'Krish Naik', 35, 'Sr. Developer')"
)

cursor.execute(
    "INSERT INTO EMPLOYEES (NAME, AGE, DEPARTMENT) VALUES('Harkirat Singh', 28, 'Web Designer')"
)

cursor.execute(
    "INSERT INTO EMPLOYEES (NAME, AGE, DEPARTMENT) VALUES('Raj Vikramaditya', 32, 'Manager')"
)

connection.commit()
```

### Showing Data from Table

```python
cursor.execute("SELECT * FROM EMPLOYEES")
rows = cursor.fetchall()

for row in rows:
    print(row)

connection.commit()
```
