Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides 2 primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (row and columns)

```python
import pandas as pd
```

### Series
A Pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to column in a table

```python
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print("Series: \n", series)
print(type(series))
```

#### Create a Series from dictionary

```python
data = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}
series_dict = pd.Series(data)
print("Series: \n", series_dict)
print(type(series_dict))
```

#### Customized Index

```python
data = [10, 20, 30]
index = ['a', 'b', 'c']
series_ind = pd.Series(data, index = index)
print("Series: \n", series_ind)
```

### DataFrame
A DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (row and columns)

#### Create a DataFrame from a dictionary of a list

```python
data = {
    'Name' : ['Krish', 'Avneesh', 'John'],
    'Age' : [25, 20, 30],
    'City' : ['Bangalore', 'Noida', 'New York']
}

df = pd.DataFrame(data)
df
```

```python
print(type(df))
```

#### Create a DataFrame from a list of dictionaries

```python
data = [
    {'Name' : 'Krish', 'Age' : 25, 'City' : 'Bangalore'},
    {'Name' : 'Avneesh', 'Age' : 20, 'City' : 'Noida'},
    {'Name' : 'John', 'Age' : 30, 'City' : 'New York'}
]

df1 = pd.DataFrame(data)
df1
```

```python
print(type(df1))
```

```python
data = [
    {'Name' : 'Krish', 'Age' : 25, 'City' : 'Bangalore'},
    {'Name' : 'Avneesh', 'Age' : 20, 'City' : 'Noida', 'Pincode' : 201306},
    {'Name' : 'John', 'Age' : 30, 'City' : 'New York'}
]

df2 = pd.DataFrame(data)
df2
```

```python
print(type(df2))
```

#### Data Manipulation with DataFrame

#####     Adding A Column

```python
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
df
```

##### Removing A Column

```python
df.drop('TotalAmount', axis = 1)
```

_Temporary Deletion of Column using only axis. From above it can be seen that the column is dropped but after printing csv again it appears once again._

```python
df
```

Solution: _Use inplace = True for permanent deletion_

```python
df.drop('TotalAmount', axis = 1, inplace=True)
```

```python
df
```

```python
df['Quantity'] = df['Quantity'] + 1
```

```python
df
```
