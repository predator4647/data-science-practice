# Data Visualization with Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them

```python
import seaborn as sns
```

Basic Plotting with Seaborn

```python
tips = sns.load_dataset('tips')
tips
```

## Scatter plot

```python
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

## Line Plot

```python
sns.lineplot(x = 'size', y = 'total_bill', data = tips)
plt.title("Line Plot of Total Bill vs Size")
plt.show()
```

## Categorical Plots

### Bar Plot

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Bar Plot of Total Bill vs Day")
plt.show()
```

### Box Plot

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Box Plot of Total Bill vs Day")
plt.show()
```

### Violin Plot

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Violin Plot of Total Bill vs Day")
plt.show()
```

## Histogram

```python
sns.histplot(tips['total_bill'], bins = 10, kde = True)
```

## KDE  Plot

```python
sns.kdeplot(tips['total_bill'], fill = True)
```

## Pair Plot

```python
sns.pairplot(tips)
```

## Heat Map

```python
correlation = tips[['total_bill', 'tip', 'size']].corr()
correlation
```

```python
sns.heatmap(correlation, annot=True, cmap='coolwarm')
```

```python
import pandas as pd
sales_df = pd.read_csv('transaction_data_100_entries.csv')
sales_df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Product Category', y='Total Revenue', data=sales_df, estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales\n(in Dollars)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Total Revenue', data=sales_df, estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Product')
plt.ylabel('Total Sales\n(in Dollars)')
plt.show()
```
