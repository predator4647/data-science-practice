It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot

### Creating a Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
```

```python
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Basic Line Plotting")
```

### Creating a Customized Line Plot

```python
plt.plot(x, y, color='red', linestyle='-')
```

```python
plt.plot(x, y, color='violet', linestyle='--')
```

```python
plt.plot(x, y, color='green', linestyle='-.', marker='o')
```

```python
plt.plot(x, y, color='blue', linestyle=':', marker='o', linewidth=2)
```

```python
plt.plot(x, y, color='blue', linestyle=':', marker='o', linewidth=4)
```

```python
plt.plot(x, y, color='blue', linestyle=':', marker='o', linewidth=4, markersize=9)
```

```python
plt.plot(x, y, color='blue', linestyle=':', marker='o', linewidth=4, markersize=9)
plt.grid(True)
```

### Multiple Plots

```python
## Sample Data
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([1, 4, 9, 16, 25])
y2 = np.array([1, 2, 3, 4, 5])

plt.figure(figsize=(9, 5))
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='green', linestyle='--')
plt.title("Plot 1")

plt.subplot(2, 2, 2)
plt.plot(y1, x, color='red', linestyle='--')
plt.title("Plot 2")

plt.subplot(2, 2, 3)
plt.plot(x, y2, color='blue', linestyle='--')
plt.title("Plot 3")

plt.subplot(2, 2, 4)
plt.plot(x**2, y2, color='violet', linestyle='--')
plt.title("Plot 4")
```

## Bar Plot

```python
categories = np.array(['A', 'B', 'C', 'D', 'E'])
values = np.array([5, 7, 3, 8, 6])
```

### Create A Bar Plot

```python
plt.bar(categories, values, color = 'PURPLE')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot")
plt.show()
```

## Histogram
Histograms are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin

```python
data = np.array([1, 1, 1, 2, 3, 3, 3, 3, 3, 4, 4, 5, 5, 5, 5])
```

### Creating a Histogram

```python
plt.hist(data, bins=5, color='orange', edgecolor='black')
```

## Pie Chart

```python
labels = np.array(['A', 'B', 'C', 'D'])
sizes = np.array([30, 20, 40, 10])
colors = np.array(['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
explode = (0.2, 0, 0, 0)
```

```python
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True)
```

_explode_: moves out the part of pie, by given difference

```python
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True)
```

## Sales Data Visualization

```python
import pandas as pd
sales_data_df = pd.read_csv('transaction_data_100_entries.csv')
sales_data_df.head()
```

```python
sales_data_df.info()
```

### Checking Sales in Each Category

```python
total_sales_by_product = sales_data_df.groupby('Product Category')['Total Revenue'].sum()
total_sales_by_product
```

```python
total_sales_by_product.plot(kind = 'bar', color = 'teal')
```

### Plot Sales Trend Over TIme

```python
sales_trend = sales_data_df.groupby('Date')['Total Revenue'].sum().reset_index()
sales_trend
```

```python
plt.plot(sales_trend['Date'], sales_trend['Total Revenue'], linestyle='--')
```
